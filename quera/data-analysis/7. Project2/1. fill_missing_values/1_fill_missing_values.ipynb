Ш§ШІ ШўЩҶШ¬Ш§ Ъ©ЩҮ ШҜШұ Ш§ЫҢЩҶ Щ…Ш¬Щ…ЩҲШ№ЩҮвҖҢШҜШ§ШҜЩҮ Ъ©Ш§ШұШ§Ъ©ШӘШұЩҮШ§ЫҢ ЪҶЫҢЩҶЫҢ ЩҲШ¬ЩҲШҜ ШҜШ§ШұШҜ ЩҲ Ш§ЩҶЪ©ЩҲШҜШұ (Encoder) ЩҫЫҢШҙвҖҢЩҒШұШ¶ ЩҫШ§ЩҶШҜШ§ШІ ЩӮШ§ШҜШұ ШЁЩҮ ШҙЩҶШ§ШіШ§ЫҢЫҢ ШўЩҶвҖҢЩҮШ§ ЩҶЫҢШіШӘ ЩҶЫҢШ§ШІ Ш§ШіШӘ ШӘШ§ ШөШұШ§ШӯШӘШ§ЩӢ Ш§ЩҶЪ©ЩҲШҜШұ Щ…ЩҶШ§ШіШЁ Ш§ЫҢЩҶ Ъ©Ш§Шұ ШұШ§ Щ…ШҙШ®Шө Ъ©ЩҶЫҢЩ…. 
 Щ„ЫҢШіШӘ Ъ©Ш§Щ…Щ„ Ш§ЩҶЪ©ЩҲШҜШұвҖҢЩҮШ§ЫҢ ЩӮШ§ШЁЩ„ Ш§ШіШӘЩҒШ§ШҜЩҮ ШҜШұ ЩҫШ§ЫҢШӘЩҲЩҶ ШҜШұ <a href=" https://docs.python.org/3/library/codecs.html#standard-encodings"> Ш§ЫҢЩҶ Щ„ЫҢЩҶЪ© </a> ШўЩ…ШҜЩҮ Ш§ШіШӘ.
ШЁШ§ Ш§ЫҢЩҶ ШӯШ§Щ„ШҢвҖҢ Ш§ЩҶЪ©ЩҲШҜШұЫҢ Ъ©ЩҮ Щ…Ш§ ШЁЩҮвҖҢШҜЩҶШЁШ§Щ„ ШўЩҶ ЩҮШіШӘЫҢЩ… ЩҲ Ш§ШІ Ъ©Ш§ШұШ§Ъ©ШӘШұЩҮШ§ЫҢ ЪҶЫҢЩҶЫҢ ЩҫШҙШӘЫҢШЁШ§ЩҶЫҢ Щ…ЫҢвҖҢЪ©ЩҶШҜШҢ <code>gbk</code> ЩҶШ§Щ… ШҜШ§ШұШҜ.
