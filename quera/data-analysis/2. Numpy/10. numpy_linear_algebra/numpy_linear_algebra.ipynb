```python
import numpy as np

A = np.array([[30, 120, 205],
              [2, 25, 4],
              [5, 0, 1]])

B = np.array([800, 50, 15])

servings = np.linalg.solve(A, B)

print("Recommended servings per day:")
print("Broccoli (X):", servings[0])
print("Chicken (Y):", servings[1])
print("Rice (Z):", servings[2])
```
