| ----------- | ----------- |
| func        | ШӘШ§ШЁШ№ЫҢ Ъ©ЩҮ ЩӮШөШҜ ШҜШ§ШұЫҢ ШұЩҲЫҢ ШҜЫҢШӘШ§ЩҒШұЫҢЩ… Ш§Ш№Щ…Ш§Щ„ Ъ©ЩҶЫҢЩ….                                                                                                                                                                                                                                       |
| axis        | - Ш§ЪҜШұ 0 ШЁШ§ШҙШҜ ШӘШ§ШЁШ№ ШұЩҲЫҢ ЩҮШұ ШіШӘЩҲЩҶ Ш§Ш№Щ…Ш§Щ„ Щ…ЫҢвҖҢШҙЩҲШҜ. | kwargs      | ЩҫШ§ШұШ§Щ…ШӘШұвҖҢЩҮШ§ЫҢ Ъ©Щ„ЫҢШҜЩҲШ§ЪҳЩҮвҖҢШ§ЫҢ (keyword arguments) Ъ©ЩҮ ШӘШ§ШЁШ№ func Щ…ЫҢвҖҢЩҫШ°ЫҢШұШҜ.                                                                                                                                                                                                               |
| result_type | Ш¬ЩҶШі Ш®ШұЩҲШ¬ЫҢ ШұШ§ Щ…ШҙШ®Шө Щ…ЫҢвҖҢЪ©ЩҶШҜ ЩҲ ЩҒЩӮШ· ШҜШұ ШөЩҲШұШӘЫҢ Щ…ЩҲШұШҜ ШӘЩҲШ¬ЩҮ ЩӮШұШ§Шұ Щ…ЫҢвҖҢЪҜЫҢШұШҜ Ъ©ЩҮ `axis=0вҖҚ` ШЁШ§ШҙШҜ. Щ…ЫҢвҖҢШӘЩҲШ§ЩҶШҜ ШіЩҮ Щ…ЩӮШҜШ§Шұ expandШҢ reduce ЩҲ broadcast ШұШ§ ШҜШ§ШҙШӘЩҮ ШЁШ§ШҙШҜ.                                                                                                                                    |

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Щ…Ш«Ш§Щ„ Ш§ЩҲЩ„
</font>
</h2>
<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    ШҜЫҢШӘШ§ЩҒШұЫҢЩ…ЫҢ ШЁЩҮ Ш§ШіЩ… <code>df</code> ШӘШ№ШұЫҢЩҒ Щ…ЫҢвҖҢЪ©ЩҶЫҢЩ… ЩҲ ШЁЩҮ ШҜЩҲ ШұЩҲШҙ Ш¬Щ…Ш№ Ш§Щ„Щ…Ш§ЩҶвҖҢЩҮШ§ЫҢ ЩҮШұ ШіШ·Шұ ШұШ§ Щ…ШӯШ§ШіШЁЩҮ Щ…ЫҢвҖҢЪ©ЩҶЫҢЩ….
</font>
</p>


````python
df = pd.DataFrame()
df['col_1'] = [1,4,9,16]
df['col_2'] = [25,36,49,64]
df
````

<h3 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ШұЩҲШҙ Ш§ЩҲЩ„
</font>
</h3>
<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    ШӘШ§ШЁШ№ЫҢ ШЁЩҮ Ш§ШіЩ… <code>sum_row</code> Щ…ЫҢвҖҢЩҶЩҲЫҢШіЫҢЩ… ЩҲ ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ ШӘШ§ШЁШ№ <code>apply</code> ШўЩҶ ШұШ§ ШұЩҲЫҢ ЩҮШұ ШіШ·Шұ Ш§ШІ ШҜЫҢШӘШ§ЩҒШұЫҢЩ… Ш§Ш№Щ…Ш§Щ„ Щ…ЫҢвҖҢЪ©ЩҶЫҢЩ….
</font>
</p>


````python
def sum_row (row) :
    return row['col_2'] + row['col_1']

df['sum'] = df.apply(sum_row, axis=1)
df
````

<h3 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
ШұЩҲШҙ ШҜЩҲЩ…
</font>
</h3>
<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
   ШҜШұ ШұЩҲШҙ ШҜЩҲЩ… ШЁЩҮ Ш¬Ш§ЫҢ ШӘШ№ШұЫҢЩҒ ЩҲ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ ШӘЩҲШ§ШЁШ№ Ш§Ш¶Ш§ЩҒЩҮШҢ Щ…ЫҢвҖҢШ®ЩҲШ§ЩҮЫҢЩ… ЩҒЩӮШ· ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ Ш№ШЁШ§ШұШӘвҖҢЩҮШ§ЫҢ Щ„Ш§Щ…ШҜШ§ Ш¬Щ…Ш№ Ш§Щ„Щ…Ш§ЩҶвҖҢЩҮШ§ЫҢ ЩҮШұ ШіШ·Шұ ШұШ§ ШЁЩҮвҖҢШҜШіШӘ ШЁЫҢШ§ЩҲШұЫҢЩ….
    <br>
    Ш§ШЁШӘШҜШ§ Щ„Ш§ШІЩ… Ш§ШіШӘ ШіШӘЩҲЩҶ <code>sum</code> Ъ©ЩҮ Ш§ШІ ШұЩҲШҙ Ш§ЩҲЩ„ Ш§Ш¶Ш§ЩҒЩҮ ШҙШҜЩҮ ШЁЩҲШҜ ШұШ§ ШӯШ°ЩҒ Ъ©ЩҶЫҢЩ….
</font>
</p>


````python
df.drop(['sum'], axis=1, inplace=True)
df['sum'] = df.apply(lambda row : row['col_1'] + row['col_2'], axis=1)
df
````

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Щ…Ш«Ш§Щ„ ШҜЩҲЩ…
</font>
</h2>
<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    ШҜШұ Щ…Ш«Ш§Щ„ ШҜЩҲЩ… ЩӮШөШҜ ШҜШ§ШұЫҢЩ… Ш¬Щ…Ш№ Ш§Щ„Щ…Ш§ЩҶвҖҢЩҮШ§ЫҢ ЩҮШұ ШіШӘЩҲЩҶ  <code>df</code> Ъ©ЩҮ ШҜШұ Щ…Ш«Ш§Щ„ ЩӮШЁЩ„ЫҢ Щ…Ш№ШұЩҒЫҢ ШҙШҜ ШұШ§ ШӯШіШ§ШЁ Ъ©ЩҶЫҢЩ….
    
<br>
    ШЁЩҮ Ш№ЩҶЩҲШ§ЩҶ ШӘШ§ШЁШ№ЫҢ Ъ©ЩҮ ЩӮШөШҜ ШҜШ§ШұЫҢЩ… ШұЩҲЫҢ ШҜЫҢШӘШ§ЩҒШұЫҢЩ… Ш§Ш№Щ…Ш§Щ„ Ъ©ЩҶЫҢЩ…ШҢ Ш§ШІ <code>sum</code> Ъ©ЩҮ ШҜШұ ЩҶШ§Щ…ЩҫШ§ЫҢ ЩҲШ¬ЩҲШҜ ШҜШ§ШұШҜ Ш§ШіШӘЩҒШ§ШҜЩҮ Ъ©ШұШҜЩҮвҖҢШ§ЫҢЩ….
</font>
</p>


````python
df.drop(['sum'], axis=1, inplace=True)
df.apply(np.sum)
````

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
ШҜШұ ШұЩҲШҙ Ш§ЩҲЩ„ Щ…ЫҢвҖҢШ®ЩҲШ§ЩҮЫҢЩ… ШЁЩҮ ШӘШұШӘЫҢШЁ ШіШ·ШұЩҮШ§ ШӘЩӮШіЫҢЩ… ШұШ§ Ш§ЩҶШ¬Ш§Щ… ШҜЩҮЫҢЩ…. ЫҢШ№ЩҶЫҢ Ш§ШЁШӘШҜШ§ ЩҮЩ…ЩҮ Ш§Щ„Щ…Ш§ЩҶвҖҢЩҮШ§ЫҢ ШіШ·Шұ Ш§ЩҲЩ„ШҢ ШіЩҫШі ЩҮЩ…ЩҮ Ш§Щ„Щ…Ш§ЩҶвҖҢЩҮШ§ЫҢ ШіШ·Шұ ШҜЩҲЩ… ЩҲ Ш§Щ„ЫҢ ШўШ®Шұ ШұШ§ ШЁШұ ЫІ ШӘЩӮШіЫҢЩ… Ъ©ЩҶЫҢЩ….
    <br>
    ШЁЩҮ Ш§ЫҢЩҶ Щ…ЩҶШёЩҲШұ ШӘШ§ШЁШ№ <code>row_divide_2</code> ШұШ§ ШӘШ№ШұЫҢЩҒ Щ…ЫҢвҖҢЪ©ЩҶЫҢЩ… ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ <code>apply</code> ШўЩҶ ШұШ§ ШұЩҲЫҢ ШіШ·ШұвҖҢЩҮШ§ЫҢ ШҜЫҢШӘШ§ЩҒШұЫҢЩ… Ш§Ш№Щ…Ш§Щ„ Щ…ЫҢвҖҢЪ©ЩҶЫҢЩ….
</font>

````python
def row_divide_2 (row) :
  return row / 2 
df.apply(row_divide_2, axis=1)
````

    Ш§ЫҢЩҶ ШЁШ§Шұ ШЁШ§ ШӘШ№ШұЫҢЩҒ ШӘШ§ШЁШ№ <code>col_divide_2</code> ЩҲ Ш§Ш№Щ…Ш§Щ„ ШіШӘЩҲЩҶ ШЁЩҮ ШіШӘЩҲЩҶ ШўЩҶ ШЁШ§ Ш§ШіШӘЩҒШ§ШҜЩҮ Ш§ШІ <code>apply</code> ЩӮШөШҜ ШҜШ§ШұЫҢЩ… ЩҮЩ…ЩҮ Ш§Щ„Щ…Ш§ЩҶвҖҢЩҮШ§ЫҢ ШҜЫҢШӘШ§ЩҒШұЫҢЩ… ШұШ§ ШЁШұ ЫІ ШӘЩӮШіЫҢЩ… Ъ©ЩҶЫҢЩ….

````python
def col_divide_2 (col) : 
    return col / 2 

df.apply(col_divide_2)
````
