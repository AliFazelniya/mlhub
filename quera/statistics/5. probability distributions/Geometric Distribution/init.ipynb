```python
from scipy.stats import hypergeom

# پارامترهای مسئله
N1 = 30  # تعداد توپ‌های آبی
N2 = 20  # تعداد توپ‌های قرمز
N = N1 + N2  # تعداد کل توپ‌ها
n = 10  # تعداد توپ‌های انتخاب‌شده
k = 4  # تعداد توپ‌های قرمز مورد نظر

# محاسبه احتمال
prob = hypergeom.pmf(k, N, N2, n)
print(f"P(X = {k}) = {prob:.4f}")
```

```python
# تولید مقادیر تصادفی از توزیع فوق هندسی
random_samples = hypergeom.rvs(N, N2, n, size=1000)

# نمایش تعداد توپ‌های قرمز در نمونه‌های تصادفی
print(random_samples[:50])
```

```python
import matplotlib.pyplot as plt

# مقدارهای ممکن برای X
x = range(0, n+1)

# محاسبه PMF برای تمام مقادیر ممکن
pmf_vals = hypergeom.pmf(x, N, N2, n)

# رسم نمودار
plt.bar(x, pmf_vals, color='red', alpha=0.7)
plt.xlabel('Number of red balls (X)')
plt.ylabel('Probability')
plt.title('Hypergeometric Distribution PMF')
plt.show()
```
