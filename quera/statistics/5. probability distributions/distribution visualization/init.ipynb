```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Generate sample data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # Normal distribution

# Plot histogram
fig = px.histogram(x=data, nbins=30, title='Histogram of Data Distribution', labels={'x': 'Values'})
fig.show()
```

```python
fig = px.box(y=data, title='Box Plot of Data Distribution')
fig.show()
```

```python
fig = px.violin(y=data, box=True, points='all', title='Violin Plot of Data Distribution')
fig.show()
```

```python
import plotly.figure_factory as ff

fig = ff.create_distplot([data], group_labels=['Data'], show_hist=False, show_rug=True)
fig.update_layout(title='Probability Density Function (PDF) Plot')
fig.show()
```

```python
from scipy.stats import skew
print("Skewness of the data:", skew(data))
```

```python
from scipy.stats import kurtosis
print("Kurtosis of the data:", kurtosis(data))
```

```python
import plotly.express as px
import pandas as pd

# خواندن مجموعه داده
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.info()
```

```python
# حذف مقادیر گمشده در ستون سن و فیلتر کردن افراد نجات یافته در داده
titanic_cleaned = titanic.dropna(subset=['age'])
titanic_survived = titanic_cleaned[titanic_cleaned['survived']== 1]
```

```python
# مصورسازی توزیع سن مسافران با هیستوگرام و منحنی چگالی (KDE)
fig = px.histogram(titanic_survived, x="age", nbins=30, marginal="rug", 
                   title="Age Distribution of Survived Titanic Passengers by Gender",
                   labels={"age": "Age", "count": "Frequency"},
                   opacity=0.7,  color="sex", barmode="overlay")

# اضافه کردن منحنی چگالی (KDE)
fig.update_traces(opacity=0.7, selector=dict(type='histogram'))
fig.add_vline(x=titanic_survived['age'].mean(), line_dash="dash", line_color="red", 
              annotation_text=f"age mean : {titanic_survived['age'].mean():.1f} years old")

# نمایش نمودار
fig.show()
```
