```python
from scipy.stats import binom
print(binom.pmf(18, 20, 0.25))
```

```python
answers_to_win = [18,19,20] # تعداد پاسخ‌های درستی که یک فرد برای برنده شدن نیاز دارد
p_win = 0 # در ابتدا این مقدار را برابر 0 قرار داده تا هر احتمالی که محاسبه میکنیم به آن بیافراییم
for k in answers_to_win:
    p_win += binom.pmf(k, 20, 0.25)
print(p_win,':شانس برنده شدن برای یک  نفر')
```
