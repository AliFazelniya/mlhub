```python
from scipy import stats
```

```python
from scipy.stats import uniform

data = uniform.rvs(size=10)

print("داده‌های تصادفی:", data)
```

```python
import numpy as np

data = [2, 4, 6, 8, 10]
mean = np.mean(data)
std = np.std(data)

print("میانگین:", mean)
print("انحراف معیار:", std)
```

```python
from scipy.stats import ttest_1samp

data = [5, 7, 8, 6, 9]
t_stat, p_value = ttest_1samp(data, 7)

print("مقدار p:", p_value)
```

```python
from scipy.stats import norm

data = norm.rvs(size=10, loc=5, scale=2) 

print("داده‌ها:", data)
```
