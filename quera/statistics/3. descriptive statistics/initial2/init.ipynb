```python
import pandas as pd
import numpy as np

df = pd.DataFrame([['Laptop', 'Electronics', 50, 800],
       ['Coffee Maker', 'Appliances', 20, 120],
       ['T-shirt', 'Clothing', 150, 25],
       ['Headphones', 'Electronics', 80, 150],
       ['Microwave', 'Appliances', 30, 200],
       ['Jeans', 'Clothing', 100, 45],
       ['Toaster', 'Appliances', 60, 80]],
       index = ['a','b','c','d','e','f','g'],
       columns= ['product' , 'category' , 'stock', 'price'])
df
```

```python
# محاسبه میانگین انحراف مطلق با pandas
mad_pandas = (df['price'] - df['price'].mean()).abs().mean()
mad_pandas
```

```python
# محاسبه میانگین انحراف مطلق با numpy
mad_numpy = np.mean(np.abs(df['price'] - np.mean(df['price'])))
mad_numpy
```

```python
# محاسبه انحراف معیار (Standard Deviation) با pandas
std_dev_pandas = df['price'].std(ddof=1)
std_dev_pandas
```

```python
# محاسبه انحراف معیار با numpy
std_dev_numpy = np.std(df['price'], ddof=1)
std_dev_numpy
```

```python
# محاسبه واریانس با pandas
variance_pandas = df['price'].var(ddof=1)
variance_pandas
```

```python
# محاسبه واریانس با numpy
variance_numpy = np.var(df['price'], ddof=1)
variance_numpy
```
