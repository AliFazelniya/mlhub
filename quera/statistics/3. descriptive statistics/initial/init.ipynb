```python
import pandas as pd
import numpy as np

df = pd.DataFrame([['Laptop', 'Electronics', 50, 800],
       ['Coffee Maker', 'Appliances', 20, 120],
       ['T-shirt', 'Clothing', 150, 25],
       ['Headphones', 'Electronics', 80, 150],
       ['Microwave', 'Appliances', 30, 200],
       ['Jeans', 'Clothing', 100, 45],
       ['Toaster', 'Appliances', 60, 80]],
       index = ['a','b','c','d','e','f','g'],
       columns= ['product' , 'category' , 'stock', 'price'])
df
```

```python
# محاسبه میانگین قیمت‌ها با pandas
mean_price = df['price'].mean()
mean_price
```

```python
# محاسبه میانگین قیمت‌ها با Numpy
mean_price_numpy = np.mean(df['price'].values)
mean_price_numpy
```

```python
# محاسبه میانگین اصلاح‌شده با pandas
trimmed_mean = df['price'].sort_values().iloc[2:-2].mean()
trimmed_mean
```

```python
# محاسبه میانگین اصلاح‌شده با Numpy
trimmed_mean_numpy = np.mean(np.sort(df['price'].values)[2:-2])
trimmed_mean_numpy
```

```python
# محاسبه میانگین وزنی با pandas
weights = [0.1, 0.1, 0.1, 0.1, 0.2, 0.2, 0.2]  # وزن‌ها
weighted_mean = np.average(df['price'], weights=weights)
weighted_mean
```

```python
# محاسبه میانه با pandas
median_value = df['price'].median()
median_value
```

```python
# محاسبه میانه با Numpy
median_value_numpy = np.median(df['price'].values)
median_value_numpy
```

```python
# محاسبه میانه وزنی با pandas
weighted_median = df['stock'].apply(lambda x: x * 0.5).median()
weighted_median
```

```python
# محاسبه میانه وزنی با Numpy
weights = np.array([0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5])
weighted_median_numpy = np.median(np.multiply(df['stock'].values, weights))
weighted_median_numpy
```

```python
# محاسبه مد با pandas
mode_value = df['stock'].mode()[0]
mode_value
```

```python
# محاسبه مد با Numpy
mode_value_numpy = np.bincount(df['stock'].values).argmax()
mode_value_numpy
```
