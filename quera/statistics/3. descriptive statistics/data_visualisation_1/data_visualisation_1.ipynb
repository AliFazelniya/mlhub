```python
import plotly.express as px


data = {'روز': ['شنبه', 'یکشنبه', 'دوشنبه', 'سه‌شنبه', 'چهارشنبه'],'فروش': [100, 150, 130, 170, 200]}  # داده نمونه


fig = px.line(data, x='روز', y='فروش', title='نمودار فروش روزانه', labels={'روز': 'روز هفته', 'فروش': 'مقدار فروش'}) # رسم نمودار خطی
fig.show()
```

```python
import pandas as pd


data = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [[10, 20, 25, 30, 35, 40], [15, 18, 22, 28, 35], [5, 10, 15, 20, 25]]}) # داده نمونه


fig = px.box(data.explode('Values'), x='Category', y='Values', title='نمودار جعبه‌ای') # رسم نمودار جعبه‌ای
fig.show()
```

```python
import numpy as np

# داده نمونه
data = np.random.normal(loc=50, scale=10, size=1000)  # تولید داده نرمال


fig = px.histogram(data, nbins=20, title='هیستوگرام توزیع داده‌ها', labels={'value': 'مقدار', 'count': 'تعداد'}) # رسم هیستوگرام
fig.show()
```

```python
# تولید جدول فراوانی
bins = np.histogram_bin_edges(data, bins=10)
freq_table = pd.cut(data, bins=bins).value_counts().sort_index()

# نمایش جدول
print(freq_table)
```

```python
fig = px.density_contour(x=data, title='نمودار چگالی توزیع داده‌ها', labels={'x': 'مقدار'}) # رسم نمودار چگالی
fig.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(1000)
# تولید داده‌های نمونه
data_class_A = np.random.normal(170, 10, 200)  # قد دانش‌آموزان کلاس A
data_class_B = np.random.normal(160, 8, 200)   # قد دانش‌آموزان کلاس B

# تبدیل داده‌ها به DataFrame
df = pd.DataFrame({
    'قد': np.concatenate([data_class_A, data_class_B]),
    'کلاس': ['A'] * 200 + ['B'] * 200
})

# رسم نمودار چگالی
fig = px.density_contour(df, x="قد", color="کلاس", 
                         title="مقایسه توزیع قد در دو کلاس",
                         labels={"قد": "قد (سانتی‌متر)"})

fig.show()
```
