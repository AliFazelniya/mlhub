```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
data = {
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'] * 100,  # 500 records
    'Employment Status': ['Employed', 'Employed', 'Unemployed', 'Unemployed', 'Employed', 'Unemployed'] * 100
}

df = pd.DataFrame(data)

# Create the contingency table
contingency_table = pd.crosstab(df['Gender'], df['Employment Status'])

# Perform the chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table, True)

# Display the results
print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected table:")
print(expected)

# Analyze the results
if p < 0.05:
    print("Result: Null hypothesis is rejected. There is a dependency between gender and employment status.")
else:
    print("Result: Null hypothesis is not rejected. There is no dependency between gender and employment status.")
```
