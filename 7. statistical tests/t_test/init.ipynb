```python
from scipy.stats import ttest_1samp

ttest_res = ttest_1samp(data,30)
print("t-statistic is: "+str(ttest_res[0][0]))
print("p-value is: "+str(ttest_res[1][0]/2)) # Division by two is needed because our test is one-way
```
