```python
import zipfile
import joblib

joblib.dump(r_value_ad, "r_value_ad")
joblib.dump(r_value_price, "r_value_price")
joblib.dump(choice, "choice")
joblib.dump(cups_per_day, "cups_per_day")


def compress(file_names):
    print("File Paths:")
    print(file_names)
    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED
    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = ["r_value_ad", "r_value_price", "choice", "cups_per_day", "init.ipynb"]
compress(file_names)
```
