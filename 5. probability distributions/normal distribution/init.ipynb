```python
from scipy.stats import norm
percent = norm.cdf(160,170,10)*100
print(f"{percent:.2f}")
```

```python
f_a = norm.cdf(160,170,10)
f_b = norm.cdf(180,170,10)
percent = (f_b - f_a)*100
print(f"{percent:.2f}")
```

```python
f = norm.cdf(185,170,10)
percent = (1- f)*100
print(f"{percent:.2f}")
```
