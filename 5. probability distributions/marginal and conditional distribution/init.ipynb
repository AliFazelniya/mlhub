```python
import numpy as np
import pandas as pd
import plotly.figure_factory as ff
import plotly.express as px
import plotly.graph_objects as go
from scipy.stats import gaussian_kde, norm
```

```python
tips = pd.read_csv('tips.csv')
```

```python
fig = px.density_contour(tips, x="total_bill", y="tip", title="Joint Distribution of Total Bill and Tip")
fig.update_traces(contours_coloring="fill", colorscale="Blues")
fig.show()
```

```python
fig = px.histogram(tips, x="total_bill", nbins=20, marginal="rug", title="Marginal Distribution of Total Bill")
fig.show()
```

```python
fig = px.histogram(tips, x="tip", nbins=20, marginal="rug", title="Marginal Distribution of Tip")
fig.show()
```

```python
# Select total_bill between 10 and 20
conditional_data = tips[(tips["total_bill"] >= 10) & (tips["total_bill"] <= 20)]["tip"]

# Fit normal distribution
mu, sigma = norm.fit(conditional_data)
x = np.linspace(min(conditional_data), max(conditional_data), 100)
pdf = norm.pdf(x, mu, sigma)

# Plot conditional histogram
fig = px.histogram(conditional_data, nbins=15, histnorm='probability density', title="Conditional Distribution of Tip Given Total Bill in [10,20]")
fig.add_scatter(x=x, y=pdf, mode='lines', name=f'Normal Fit: μ={mu:.2f}, σ={sigma:.2f}', line=dict(color='red', dash='dash'))
fig.show()
```
